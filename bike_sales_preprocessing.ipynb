# Preprocessing - Tiền xử lý dữ liệu

## Xoá dữ liệu trùng lặp

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python
df.ID.value_counts()
```

Dữ liệu đã giảm xuống từ 1026 thành 1000 dòng (số 1000 đúng bằng thống kê số lượng dữ liệu khác nhau của cột ID). Như vậy vấn đề ID trùng nhau là do dữ liệu bị lặp lại.

## Chuyển cột Income về định dạng dữ liệu Int

```python
df['Income'] = (
    df['Income']
    .str.replace('$', '', regex=False)     # bỏ ký hiệu $
    .str.replace('.', '', regex=False)     # bỏ dấu chấm (phân cách nghìn)
    .str.replace(',', '.', regex=False)    # đổi dấu phẩy thành dấu chấm (thập phân)
    .astype(float)                         # chuyển sang float
    .astype(int)                           # chuyển sang int
)
```

```python
df.Income.value_counts()
```

## Sửa lại giá trị cho cột Marital Status

```python
df['Marital Status'] = df['Marital Status'].map({
    'M': 'Married',
    'S': 'Single'
})
```

```python
df.head()
```

## Sửa lại giá trị cho cột Gender

```python
df['Gender'] = df['Gender'].replace({
    'M': 'Male',
    'F': 'Female'
})
```

```python
df.head()
```

## Tạo cột phân cụm tuổi tác

```python
df['Age Group'] = df['Age'].apply(
    lambda x: 'Adult(0-30)' if x <= 30
    else 'Middle Age(31-54)' if x <= 54
    else 'Old(54+)'
)
```

## Thay đổi giá trị trong cột Commute Distance

```python
df['Commute Distance'] = df['Commute Distance'].replace(
    '10+ Miles', 'More than 10 miles '
)
```
